### Задача 2  Численное интегрирование (4 балла)

Вычислить интеграл

$$
\int_{1.0}^{2.6} e^{-0.1 / x} d x
$$

используя квадратурные формулы:

- центральных прямоугольников с шагом $h=0.4$; дать __априорную__ оценку погрешности;

- трапеций с шагами $h=0.4$ и $h=0.2$; оценить погрешность результата по формуле Рунге и уточнить результат по Рунге;

- Симпсона с шагом $h=0.4$; дать __априорную__ оценку погрешности.

- Построить в логарифмических координатах график зависимости ошибки от $h$ для формулы Симпсона, взяв $h = 10^{-k}, k = 1, 2, ... 12$. Ошибку оценивать по формуле Рунге ($p = 4$).
$$
\frac{\left|y_{i, h}-y_{i, h / 2}\right|}{2^p-1}  
$$

p - порядок точности использованного численного метода.
