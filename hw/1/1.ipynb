1) Пусть у двух матриц совпадает количество строк. Объедините их в одну горизонтально (объедините столбцы этих матриц)

```python
n = 3
m1 = 4
m2 = 5

M = np.zeros((n, m1))
N = np.ones((n, m2))

C = np.hstack((M, N))
print(C)
```

2) Пусть дана numpy-матрица
и два массива индексов
одинаковой длины. Постройте массив элементов
Указание: для выполнения этой операции не требуется никаких именных функций.

```python
n = 4
a = np.array([1, 2, 3, 4]) - 1
b = np.array([4, 3, 2, 1]) - 1
M = np.zeros((n, n))


for i in range(n*n):
    M[i//n][i%n] = i
    
c = np.zeros(n)
for i in range(n):
    c[i] = M[a[i]][b[i]]
    
print(c)
```

3) Подсчитайте след матрицы и количество ненулевых элементов на ее диагонали

```python
n = 3
m = 4
M = np.ones((n, m))

print(M.trace())

count = np.sum(np.array(M[i][i] != 0)for i in range(n))
print(count)
```

4) Даны два произвольных numpy-массива одинаковой размерности (n,m)
. Замените нулями те элементы второго массива, для которых на соответствующем месте в первом массиве находятся отрицательные числа

```python
n = 2
m = 3

array1 = np.random.randint(-10, 10, size=(n, m))
array2 = np.random.randint(-10, 10, size=(n, m))

mask = array1 < 0
array2[mask] = 0

print(array1)
print(array2)
```

5) Даны две матрицы с одинаковым количеством строк и количеством столбцов
и
соответственно. Постройте матрицу mxm
всевозможных произведений столбцов друг на друга (в смысле скалярного произведения).

```python
n = 3
m = 4
matrix1 = np.random.randint(0, 10, size=(n, m))
matrix2 = np.random.randint(0, 10, size=(n, m))

result = np.dot(matrix1.T, matrix2)  # каждый элемент (i, j) будет скалярным произведением i-го столбца первой матрицы на j-й столбец второй матрицы.

print("Матрица 1:\n", matrix1)
print("Матрица 2:\n", matrix2)
print("Результат:\n", result)
```

## ЗАДАНИЕ 2

```python
import matplotlib.pyplot as plt
import numpy as np
```

Постройте таблицу 2 x 2
из четырех графиков (см. функцию plt.subplots)

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2
y4 = np.sqrt(x)

fig, axs = plt.subplots(2, 2, figsize=(8, 6))

axs[0, 0].plot(x, y1)
axs[0, 0].set_title('График 1: sin(x)')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('График 2: cos(x)')

axs[1, 0].plot(x, y3)
axs[1, 0].set_title('График 3: x^2')

axs[1, 1].plot(x, y4)
axs[1, 1].set_title('График 4: sqrt(x)')

plt.tight_layout()
plt.show()
```

Постройте график с крестами ошибок

```python
x = np.arange(1, 11)
y = np.random.rand(10) * 10
errors = np.random.rand(10) * 3

plt.errorbar(x, y, yerr=errors, fmt='o', capsize=5, linestyle='--', color='blue', markerfacecolor='red', markersize=8)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('График с крестами ошибок')

plt.show()
```

Постройте график в логарифмическом масштабе без явного вычисления логарифмов

```python
x = np.linspace(0.1, 10, 100)
y = np.exp(x)

plt.semilogy(x, y)
plt.xlabel('Значения по оси X')
plt.ylabel('Значения по оси Y (логарифмический масштаб)')
plt.title('График с логарифмическим масштабом по оси Y')
plt.grid(True)
plt.show()
```
