1) Пусть у двух матриц совпадает количество строк. Объедините их в одну горизонтально (объедините столбцы этих матриц)

2) Пусть дана numpy-матрица
и два массива индексов
одинаковой длины. Постройте массив элементов
Указание: для выполнения этой операции не требуется никаких именных функций.

```python
n = 4
a = np.array([1, 2, 3, 4]) - 1
b = np.array([4, 3, 2, 1]) - 1
M = np.zeros((n, n))


for i in range(n*n):
    M[i//n][i%n] = i
    
c = np.zeros(n)
for i in range(n):
    c[i] = M[a[i]][b[i]]
    
print(c)
```

```python
n = 3
m = 4
M = np.ones((n, m))

print(M.trace())

count = np.sum(np.array(M[i][i] != 0)for i in range(n))
print(count)
```

5) Даны две матрицы с одинаковым количеством строк и количеством столбцов
и
соответственно. Постройте матрицу mxm
всевозможных произведений столбцов друг на друга (в смысле скалярного произведения).

```python
n = 3
m = 4
matrix1 = np.random.randint(0, 10, size=(n, m))
matrix2 = np.random.randint(0, 10, size=(n, m))

result = np.dot(matrix1.T, matrix2)  # каждый элемент (i, j) будет скалярным произведением i-го столбца первой матрицы на j-й столбец второй матрицы.

print("Матрица 1:\n", matrix1)
print("Матрица 2:\n", matrix2)
print("Результат:\n", result)
```

## ЗАДАНИЕ 6

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 200
M = 50
m = 200
```

1, 2 пункты

```python
lambdas_all = []
for _ in range(M):
    matrix = np.random.normal(0, 1, (N, N))
    matrix = (matrix + matrix.T) / 2
    lambdas = np.linalg.eigvalsh(matrix)
    lambdas_all.extend(lambdas)

plt.hist(lambdas_all, bins=25)
plt.xlabel("Собственные значения")
plt.ylabel("Частота")
plt.title("Гистограмма собственных значений")
plt.show()

distances_all = []
for _ in range(m):
    matrix = np.random.normal(0, 1, (N, N))
    matrix = (matrix + matrix.T) / 2
    lambdas = np.linalg.eigvalsh(matrix)
    lambdas.sort()
    distances = np.diff(lambdas)
    distances_all.extend(distances)

plt.hist(distances_all, bins=50)
plt.xlabel("Расстояния между собственными значениями")
plt.ylabel("Частота")
plt.title("Гистограмма расстояний")
plt.show()
```

3 пункт

```python

def wigner_func_approx(rho):
  """
  с1* rho * exp(-c2 * rho^c3).
  """
  c1 = 60
  c2 = 15
  c3 = 1.5
  return c1 * rho * np.exp(-c2 * rho**c3)


plt.hist(distances_all, bins=100, density=True, label="Гистограмма")
rho_values = np.linspace(0, max(distances_all), 100)
plt.plot(rho_values, wigner_func_approx(rho_values), label="Аппроксимация")
plt.xlim(0, 1)
plt.xlabel("Расстояния между собственными значениями")
plt.ylabel("Плотность вероятности")
plt.title("Гистограмма расстояний и аппроксимация")
plt.legend()
plt.show()
```

4 пункт

```python
lambdas_all = []
for _ in range(M):
    matrix = np.random.normal(0, 1, (N, N)) + 1j * np.random.normal(0, 1, (N, N))
    matrix = (matrix + matrix.conj().T) / 2 # эрмитово сопряженная = комплексное сопряжение + транспонирование
    lambdas = np.linalg.eigvalsh(matrix)
    lambdas_all.extend(lambdas)

plt.hist(lambdas_all, bins=25)
plt.xlabel("Собственные значения")
plt.ylabel("Частота")
plt.title("Гистограмма собственных значений (эрмитова матрица)")
plt.show()

distances_all = []
for _ in range(m):
    matrix = np.random.normal(0, 1, (N, N))
    matrix = (matrix + matrix.T) / 2
    lambdas = np.linalg.eigvalsh(matrix)
    lambdas.sort()
    distances = np.diff(lambdas)
    distances_all.extend(distances)


plt.hist(distances_all, bins=100, density=True, label="Гистограмма расстояний")
rho_values = np.linspace(0, max(distances_all), 100)
plt.plot(rho_values, wigner_func_approx(rho_values), label="Аппроксимация")
plt.xlim(0, 1)
plt.xlabel("Расстояния между собственными значениями")
plt.ylabel("Частота")
plt.title("Гистограмма расстояний и аппроксимация")
plt.legend()
plt.show()
```

*Вывод* : на гистограммах с собственными значениями вырисовывается полукруг, плотность вероятности случайной величины $\rho$ получена эксперементально и имеей вид $c_1 \rho e^{-c_2 \rho^{c_3}}$
